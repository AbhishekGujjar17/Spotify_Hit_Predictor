### About the dataset
We are using [The Spotify Hit Predictor Dataset (1960-2019)](https://www.kaggle.com/datasets/theoverman/the-spotify-hit-predictor-dataset?select=dataset-of-00s.csv) dataset consisting of features for tracks from the 2000s retrieved using the Spotify web API. The tracks are labelled '1' or '0' ('Hit' or 'Flop') depending on how well they meet some criteria of the author. It's not always easy to tell whether a track is a success or a failure, but in the music industry it's often labelled as such based on how well it performs on the charts. Even if a track is a flop, it's still worth putting some effort into it because it can still be enjoyable to listeners. The term "flop" is simply used to describe a track that is not likely to be popular with the mainstream audience.


### Description of Variables
* **track:** the name of the track.

* **artist:** the name of the artist.

* **uri:** the resource identifier of the track. 

* **danceability:** a numerical measure that ranges from 0.0 to 1.0 and represents how well a track is suited for dancing based on a number of musical elements. These elements include tempo, rhythm stability, beat strength, and overall regularity. The value 0.0 represents the least danceable songs and 1.0 represents the most danceable.

* **energy:** a numerical measure that ranges from 0.0 to 1.0 and reflects our perceptions of how energetic or intense a song is. Typically, energetic tracks are fast, loud, and noisy, such as rock or metal songs. Musical elements such as dynamic range, perceived loudness, timbre, onset rate, and entropy are all perceptual features that contribute to this attribute.

* **key:** the estimated overall key of the track. The pitches are mapped to integers using standard Pitch Class notation. E.g. 0 = C, 1 = C?/D?, 2 = D, and so on.

* **loudness:** the average loudness of the track in decibels (dB). Typical values ​​range from -60dB to 0 dB.

* **mode:** indicates if the track has a major or minor scale. Minor is represented by 0 and major is represented by 1.

* **speechiness:** a numerical measure that ranges from 0.0 to 1.0 and indicates if a track contains spoken words. The closer it is to 1.0, the more speech-like the recording is. Tracks with values above 0.66 are most likely composed entirely of spoken words, such as a talk show or an audio book. Tracks that may contain both music and speech are described by values between 0.33 and 0.66, including rap music. Music and other non-speech-like tracks are most likely represented by values lower than 0.33.

* **acousticness:** a numerical confidence level that ranges from 0.0 to 1.0 and inidicates how likely it is that the track is acoustic. 1.0 indicates high confidence in the acoustic nature of the track and 0.0 indicates low confidence.

* **instrumentalness:** a numerical confidence level that ranges from 0.0 to 1.0 and indicates how likely it is that a track lacks vocals. It is intended that values above 0.5 represent instrumental tracks; however, confidence increases as the value approaches 1.0.

* **liveness:** a numerical confidence level that ranges from 0.0 to 1.0 and inidicates how likely it is that the track was performed live by identifying an audience's presence in the recording. A value greater than 0.8 indicates a high likelihood of the track being live.

* **valence:** a numerical measure that ranges from 0.0 to 1.0 and describes how positive a track is. Tracks with a low valence sound more negative, such as sad, depressed, or angry songs. Tracks with a high valence sound more positive, such as cheerful and upbeat songs.

* **tempo:** the estimated tempo of a track as a whole, expressed in beats per minute (BPM). This is the speed or pace of a piece in musical terms and it is directly related to the average beat duration.

* **duration_ms:** the duration of the track in milliseconds.

* **time_signature:** a track's estimated overall time signature. Indicates the number of beats in each bar or measure.

* **chorus_hit:** the timestamp of an estimate of when the track's chorus would begin. Data from an Audio Analysis API was used to extract this feature.

* **sections:** the number of sections included in the track. Data from an Audio Analysis API was used to extract this feature. 

* **target:** the target variable indicating if the track is a "hit" or "flop". A value of 1 indicates that this song is a "hit" because it has appeared at least once on the weekly list of Hot-100 tracks issued by Billboards during that decade. The number "0" denotes that the track is a "flop."

### Important Note - Definition of a "Flop"
The following is the dataset author's definition of a "flop" track:

* The track must not be on that decade's "hit" list.
* The artist of the song must not be on the decade's "hit" list.
* The genre that the track falls under must be one that is both unconventional and/or avant-garde. 
* A song from the track's genre must not be on the "hit" list.
* "US" must be one of the tracks' markets.

### Simulating Missing Data

Originally, our dataset had no missing or null values for any of the columns. Therefore, we created a procedure to simulate missing data by randomly setting 10% of the values for the danceability and the energy attributes to NaN.

```python
# Shows that there are no missing/null values for any of the columns in the original dataset.
df_music.info()
```

```python
import random
from typing import List

'''
Removes a given percentage of values from the given array by randomly replacing values with NaN.

@param percent_missing:    the percentage of total values to remove.
@param arr:                the list to remove the values from.
@returns                   the list with the values removed from it.
'''
def remove_vals(percent_missing, arr):

    num_missing_vals = round(len(arr) * percent_missing)
    # generate an array of unique indexes
    missing_vals_indexes = random.sample(range(len(arr)), num_missing_vals)

    # replace the values at the given indexes with NaN
    for i in missing_vals_indexes:
        arr[i] = np.nan
    
    return arr

'''
Creates copy of danceability and energy columns and call the remove_vals method to remove a certain 
percentage of values from them.

@returns lists containing the values that were removed from the danceability and energy attributes.
'''
def simulate_missing_vals():
    danceability_vals = df_music['danceability'].copy()
    energy_vals = df_music['energy'].copy()

    danceability_vals = remove_vals(0.10, danceability_vals)
    energy_vals= remove_vals(0.10, energy_vals)

    # replace the columns with the new columns with values removed
    df_music['danceability'] = danceability_vals
    df_music['energy'] = energy_vals
```

### Detecting Outliers

Initially, we checked for outliers for each attribute by checking which values were more than 3 standard deviations from the mean. Using this method, a significant number of outliers were detected. However, it is important to note that the attributes containing outliers have skewed distributions so this method of detecting outliers is not ideal for them.

Furthermore, we manually checked the most extreme of the outliers for each attribute and found that they were all true outliers. This means that they were not caused by an error and instead just represent real songs on Spotify that are unusual in some way. For example, there are several songs in the dataset that are over an hour long and therefore have a rare and extremely high value for the `duration_ms` and `sections` attributes.

```python
'''
Shows the number of outliers present in the column by finding if the difference between the value and the mean
is greater than 3 times of the standard deviation of the column.

@param attribute:     the column from which outliers will be found.
@returns              numbers of outliers in a column
'''
def count_outliers_using_mean(attribute: str):
    num_outliers = len(df_music[np.abs(df_music[attribute]-df_music[attribute].mean()) > (3*df_music[attribute].std())])
    print("The number of outliers for " + attribute + " is: " + str(num_outliers))
    return num_outliers
```

```python
# Calls the above function for each of the numerical attributes to show the number of detected outliers 
# for each

for attribute in df_music.select_dtypes(include="number"):
    count_outliers_using_mean(attribute)
```

Due to the fact that many of the attributes have a very skewed distribution and that the outliers found are true outliers, we decided to use a different method to detect the outliers that should be removed.

For the attributes for which there were outliers detected above, the songs with the most extreme outliers for the `speechiness`, `liveness`, `tempo`, and `chorus_hit` are fairly normal songs so we decided they should not be removed at this time. However, the songs with the most extreme outliers for the `duration_ms`, `time_signature`, and `sections` are very unusual so we decided that they should be removed.

In order to pick out the most extreme outliers for these attributes, we used quantile-based flooring and capping. This will allow us to remove some of the most extreme outliers that are most likely to bias the model negatively, while still preserving some of the slightly unusual songs. Keeping some of these somewhat unusual songs will help teach the model that they will generally not be hits. We used **0.005** as the lower quartile and **0.995** as the upper quartile since after some experimentation, using these values seems to detect only the very unusual songs.

```python
'''
Shows the number of outliers present in the column by finding how many of the values
in the column are within either the given upper quantile or the given lower quantile.

@param attribute:        the column from which outliers will be found.
@param lower_quantile:   the lower quantile value.
@param upper_quantile:   the upper quantile value.
'''
def count_outliers_using_quantiles(attribute: str, lower_quantile: float, upper_quantile: float):
    lower = df_music[attribute].quantile(lower_quantile)
    upper = df_music[attribute].quantile(upper_quantile)
    count_outliers_using_limits(df_music, attribute, lower, upper)

'''
Shows the number of outliers present in the column by finding how many of the values
in the column are within either above the given upper limit or below the given lower 
limit.

@param df:           the dataframe to use.
@param attribute:    the column from which outliers will be found.
@param lower_lim:    the lower limit value.
@param upper_lim:    he upper limit value.
'''
def count_outliers_using_limits(df: pd.DataFrame, attribute: str, lower_lim: float, upper_lim: float):
    num_outliers = len(df[df[attribute] > upper_lim]) + len(df[df[attribute] < lower_lim])
    print("The number of outliers for " + attribute + " is: " + str(num_outliers))
```

```python
# Calls the above function for each of the specified attributes to show the number of detected outliers for each.
# Uses 0.005 as the lower quartile and 0.995 as the upper quartile.

for attribute in ["duration_ms", "time_signature", "sections"]:
    count_outliers_using_quantiles(attribute, 0.005, 0.995)
```

For all of the extreme outliers that we detected and decided should be removed, we removed by replacing them with the appropriate upper or lower limit based on the quartile ranges we used for detecting them. In particular, we replaced the high outliers with the value corresponding to the 0.995 quartile and we replaced the low outliers with the value corresponding to the 0.005 quartile.

We also demonstrated that the outliers have been properly removed by showing that there are no more outliers outside of our defined acceptable range present in the dataset afterwards.

```python
'''
Removes the outliers present in each column of the dataset by replacing them with either the given 
upper limit or the given lower limit.

@param df:          the dataframe to use.
@param attribute:   the column from which outliers will be found.
@param lower_lim:   the lower limit value.
@param upper_lim:   the upper limit value.  
'''
def replace_outliers_with_limits(df: pd.DataFrame, attribute: str, lower_lim: float, upper_lim: float):
    df[attribute] = np.where(df[attribute] >= upper_lim, upper_lim, np.where(df[attribute] <= lower_lim, lower_lim, df[attribute]))
```

```python
# Calls the above functions for each of the specified attributes to remove the outliers and then show that
# there are no remaining outliers in the dataset.
for attribute in ["duration_ms", "time_signature", "sections"]:
    lower = df_music[attribute].quantile(0.005)
    upper = df_music[attribute].quantile(0.995)
    replace_outliers_with_limits(df_clean, attribute, lower, upper)
    count_outliers_using_limits(df_clean, attribute, lower, upper)
```

An alternative solution would be to place the track titles and/or artists into categories based on genre or some other common information and then encode the encode the categories using one of the above strategies. To explore the effects of each scaling technique presented in the course, we first plotted a comparison of the original data with each scaling technique.

```python
#import all the necessary scaling techniques required
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import FunctionTransformer

'''
Plots the histogram and line desnsity plots for scalers to see the effects of the scaling techniques on each attribute 

@param df:             the dataframe to use.
@param columns:        the columns on which scaling techniques will be applied.
@param scalers:        the list of initialiazation of the scalers to be used.
@param scaler_names:   the list of name of scalers to be used.
'''
def plot_histograms_density_for_scalers(df_, columns, fig_title, scalers, scaler_names):
    fig, axs = plt.subplots(len(columns),len(scalers)+1,figsize=(20,50),constrained_layout=True)
    fig.suptitle(fig_title, fontsize=16)
    axs = axs.flatten()
    i = 0 
    for c in columns:
        df_clean[c].hist(ax=axs[i], density=True) # normalizes the density
        df_clean[c].plot.density(ax=axs[i], title=c)
        i+=1
        for j in range(len(scalers)):
            df_transformed = scalers[j].fit_transform(df_)
            df_transformed = pd.DataFrame(df_transformed, index=df_.index, columns=columns)
            df_transformed[c].hist(ax=axs[i], density=True, stacked=True) # normalizes the density
            df_transformed[c].plot.density(ax=axs[i], title=scaler_names[j])
            i+=1

#scaling techniques choosen
scaler_names = ['MinMax', 'Z-score', 'MaxAbs', 'Robust', 'Quantile', 'Log']
#initializing the choosen scaling techniques
arr_scalers = [MinMaxScaler(), Normalizer(), MaxAbsScaler(), RobustScaler(), QuantileTransformer(n_quantiles=10, random_state=0, output_distribution='normal'), FunctionTransformer(np.log1p)]

# make a new dataframe that doesn't include the target since we don't want to scale it
df_scale = df_clean.copy().drop(columns=["target"])

#plotting the histogram to see the effects of each scaling technique
plot_histograms_density_for_scalers(df_scale, df_scale.columns, 'Scalers effect on data', arr_scalers, scaler_names)
```

Based on the plots above along with prior knowledge about these scaling techniques, we made the following observations:
* Min-max scaling has some potential benefits since it preserves the original distribution of each attribute and also produces a consistent range from [0, 1] for each of the attributes. However, it may not be the best choice for certain attributes such as the `chorus_hit` attribute due to severe skewness.
* Z-score scaling may not work well for most of our attributes since they have distributions that are far from normal. Also, we can see from the plots that for our data it leads to values with very small magnitude and ranges that vary significantly from one attribute to another. 
* Max-abs scaling has similar benefits and drawbacks to min-max scaling.
* Robust scaling preserves the distributions of the attributes. It also helps makes the ranges for each attribute more similar but they do vary more than the ranges produced by some other scaling techniques. 
* Quantile transform scaling produces fairly consistent ranges for each attributes and makes the distributions for the attributes more normal. 
* Log scaling makes the distributions closer to normal distributions which can help some models. However, one drawback is that it has no predetermined range to scale to so some attributes have significantly different ranges to each other. For example, after scaling most of the values for the `duration_ms` attribute are still > 10 while many of the other attributes have a from from [0, 1].

Therefore, based on this analysis, we decided to keep all of the scaling techniques except for z-score scaling and log scaling in our experimental design. Since the general guideline is to test several scaling techniques and then compare the performance to find the best results, it is best to include most of the scaling techniques. 

However, Z-score scaling was removed since it does not work well with many of the attributes' distributions. Log scaling was removed since even after scaling, some attributes still have very different ranges compared to each other.

Therefore, only the MinMax, MaxAbs, Robust, and Quantile transform scaling techniques are included since they produced better results with more consistent ranges for each attribute.

Therefore, it is easy to compare the R^2 score for different model and scaler combinations.

#### Filling in Missing Data and Outliers

Lastly, we used the selected combination to replace all of the missing data and outliers in our dataframe.

Since there are some rows where the `danceability` and `energy` attributes both have `NaN` values and a linear regression cannot be trained using rows with `NaN` values, we trained a multi-output model to predict both of those attributes.

Then, we trained models to replace the outliers for the `duration_ms`, `sections`, and `time_signature` attributes separately. We did this since we do not have a large number of attributes in our dataset, so having more attributes available in the training data for each model will likely lead to better performance.

```python
'''
Replaces all the missing data in the danceability and energy columns using the
selected model + scaling technique.

@returns:    the dataframe with the missing data replaced.
'''
def replace_missing_data_using_model():
    #initializing the final selected scaling technique 
    minMax = MinMaxScaler()
    #initializing the final selected model
    model = LinearRegression()

    # get all the rows with missing data since these are the rows we need to fix
    df_nan_data = df_music[df_music.isnull().any(axis=1)]

    # training data is all rows in df_music without missing data
    training_df = df_music[~df_music.isnull().any(axis=1)]
    training_df = training_df.drop(['track', 'artist'], axis=1)

    #training the model to forecast the values for the energy and danceability columns
    X = training_df.drop(['energy', 'danceability'], axis=1).to_numpy()
    X_train = minMax.fit_transform(X)
    y_train = training_df[['energy', 'danceability']].to_numpy()
    model.fit(X_train, y_train)

    # using the model to predict the missing rows
    df_nan_data = df_nan_data.drop(["track", "artist"], axis=1)
    X_pred = minMax.transform(df_nan_data.drop(['energy', 'danceability'], axis=1).to_numpy())
    y_pred = model.predict(X_pred)
    # putting the rows back with the forecasted values
    df_nan_data[['energy', 'danceability']] = y_pred

    df_final = training_df.append(df_nan_data)
    #shows that there is no missing data in the entire dataset
    df_final.info()

    return df_final

df_final = replace_missing_data_using_model()
```

```python
'''
Gets all of the rows in the given dataframe where the value for the given 
attribute is an outlier.

@param attribute:    The attribute that we are getting the outliers for.
@param df:           The dataframe to get the outliers from.
@returns:            A dataframe containing only the rows with the selected outliers.
'''
def get_attribute_outliers(attribute, df):
    lower = df[attribute].quantile(0.005)
    upper = df[attribute].quantile(0.995)
    return df[(df[attribute] > upper) | (df[attribute] < lower)]

'''
Replaces all the outliers for the given attributes using the selected 
model + scaling technique.

@param attributes:    The attributes to replace the outliers for.
@returns:             The dataframe with the outliers replaced.
'''
def replace_outliers_using_model(attributes, df_final):
    for attribute in attributes:
        #initializing the final selected scaling technique 
        minMax = MinMaxScaler()
        #initializing the final selected model
        model = LinearRegression()

        # get all the rows with outliers since these are the rows we need to fix
        df_attribute_outliers = get_attribute_outliers(attribute, df_final).copy()

        # training data is all rows in df_final without these outliers
        training_df = df_final[~df_final.index.isin(df_attribute_outliers.index)]

        #training the model to forecast the values for the energy and danceability columns
        X = training_df.drop([attribute], axis=1).to_numpy()
        X_train = minMax.fit_transform(X)
        y_train = training_df[[attribute]].to_numpy()
        model.fit(X_train, y_train)

        # using the model to predict the missing rows
        X_pred = minMax.transform(df_attribute_outliers.drop([attribute], axis=1).to_numpy())
        y_pred = model.predict(X_pred)
        # putting the rows back with the forecasted values
        df_attribute_outliers[attribute] = y_pred

        df_final = training_df.append(df_attribute_outliers)
        return df_final

attributes_with_outliers = ["duration_ms", "sections", "time_signature"]
df_final = replace_outliers_using_model(attributes_with_outliers, df_final)
```
